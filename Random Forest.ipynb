```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
from sklearn.cross_validation import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn import model_selection

#PART-1

#1.1
X_train = pd.read_table("train/X_train.txt", delim_whitespace = True)
y_train = pd.read_table("train/y_train.txt", delim_whitespace = True)
X_test = pd.read_table("test/X_test.txt", delim_whitespace = True)
y_test = pd.read_table("test/y_test.txt", delim_whitespace = True)

print("Number of features in the dataset:")

train_rows, train_cols = X_train.shape
print("Training Data has: {} rows and {} cols".format(train_rows, train_cols))

test_rows, test_cols = X_test.shape
print("Testing Data has: {} rows and {} cols".format(test_rows, test_cols))
```

```python
from sklearn.ensemble import RandomForestClassifier

a=list(range(0,3))
b=list(range(0,3))
n=list(range(0,9))
number_val = [200, 500, 700]
depth_val = [300, 500, 600]
f1_r = []

for i in a:
    for j in b:
        rfclf = RandomForestClassifier(n_estimators=number_val[i], max_depth=depth_val[j])
        kfold = model_selection.KFold(n_splits = 10, random_state = 42)
        f1 = model_selection.cross_val_score(rfclf, X_train, y_train, cv=kfold, scoring = 'f1_macro')
        f1_r.append(f1.mean())
```

```python
print ("F1 scores: {}".format(f1_r))
bk = n[f1_r.index(max(f1_r))]+1
g = np.ceil(bk/3)
g=g-1
c = np.remainder(bk,3)
if (c == 0):
    c = 2
else:
    c = c-1
print("The best value of number of trees is: ",number_val[int (g)])
print("The best value of tree depth is: ",depth_val[c])
```

```python
rfclf = RandomForestClassifier(n_estimators=number_val[int (g)], max_depth=depth_val[c])
rfclf.fit(X_train,y_train)

rfc_acc_test = metrics.accuracy_score(y_test, rfclf.predict(X_test))
rfc_f1_test = metrics.f1_score(y_test, rfclf.predict(X_test), average = 'macro')
print ("SVM Testing Accuracy: {}".format(rfc_acc_test))
print ("SVM F1 Score: {}".format(rfc_f1_test))

rcf_cm_test = metrics.confusion_matrix(y_test, rfclf.predict(X_test))

plt.clf()
plt.imshow(rcf_cm_test, interpolation='nearest', cmap=plt.cm.Wistia)
classNames = ['1', '2', '3', '4', '5', '6']
plt.title('Model Confusion Matrix')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
tick_marks = np.arange(len(classNames))
plt.xticks(tick_marks, classNames, rotation=45)
plt.yticks(tick_marks, classNames)
for i in range(len(classNames)):
    for j in range(len(classNames)):
        plt.text(j,i,str(rcf_cm_test[i][j]))
plt.show()
```
