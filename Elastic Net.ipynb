```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
from sklearn.cross_validation import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn import model_selection

#PART-1

#1.1
X_train = pd.read_table("train/X_train.txt", delim_whitespace = True)
y_train = pd.read_table("train/y_train.txt", delim_whitespace = True)
X_test = pd.read_table("test/X_test.txt", delim_whitespace = True)
y_test = pd.read_table("test/y_test.txt", delim_whitespace = True)

print("Number of features in the dataset:")

train_rows, train_cols = X_train.shape
print("Training Data has: {} rows and {} cols".format(train_rows, train_cols))

test_rows, test_cols = X_test.shape
print("Testing Data has: {} rows and {} cols".format(test_rows, test_cols))
```

```python
from sklearn.linear_model import SGDClassifier

a=list(range(0,6))
b=list(range(0,5))
n=list(range(0,30))
alpha_val = [1e-4,3e-4,1e-3,3e-3, 1e-2,3e-2]
l1_val = [0,0.15,0.5,0.7,1]
f1_e = []

for i in a:
    for j in b:
        enet = SGDClassifier(loss='log', penalty='elasticnet', alpha=alpha_val[i], l1_ratio=l1_val[j], max_iter=5)
        kfold = model_selection.KFold(n_splits = 10, random_state = 42)
        f1 = model_selection.cross_val_score(enet, X_train, y_train, cv=kfold, scoring = 'f1_macro')
        f1_e.append(f1.mean()) 

print ("F1 scores: {}".format(f1_e))
bk = n[f1_e.index(max(f1_e))]+1
av = np.ceil(bk/6)
av=av-1
lr = np.remainder(bk,5)
if (lr == 0):
    lr = 4
else:
    lr = lr-1
```

```python
print("The best value of alpha is: ",alpha_val[int(av)])
print("The best value of l1 is: ",l1_val[int (lr)])
enet = SGDClassifier(loss='log', penalty='elasticnet', alpha=alpha_val[int(av)], l1_ratio=l1_val[int (lr)], max_iter=100) 
enet.fit(X_train,y_train)

enet_acc_test = metrics.accuracy_score(y_test, enet.predict(X_test))
enet_f1_test = metrics.f1_score(y_test, enet.predict(X_test), average = 'macro')
print ("Elastic Net Testing Accuracy: {}".format(enet_acc_test))
print ("Elastic Net F1 Score: {}".format(enet_f1_test))

enet_cm_test = metrics.confusion_matrix(y_test, enet.predict(X_test))

plt.clf()
plt.imshow(enet_cm_test, interpolation='nearest', cmap=plt.cm.Wistia)
classNames = ['1', '2', '3', '4', '5', '6']
plt.title('Model Confusion Matrix')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
tick_marks = np.arange(len(classNames))
plt.xticks(tick_marks, classNames, rotation=45)
plt.yticks(tick_marks, classNames)
for i in range(len(classNames)):
    for j in range(len(classNames)):
        plt.text(j,i,str(enet_cm_test[i][j]))
plt.show()
```
